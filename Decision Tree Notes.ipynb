```python

#Boosting is another ensemble technique to create a collection of predictors. 
#In this technique, learners are learned sequentially with early learners fitting simple models to the data 
#and then analyzing data for errors. 
#In other words, we fit consecutive trees (random sample) and at every step, 
#the goal is to solve for net error from the prior tree.
```
